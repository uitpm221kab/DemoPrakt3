```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
 
 
# Create axis
axes = [5, 5, 5]
 
# Create Data
data = np.ones(axes, dtype=np.bool)
 
# Control Transparency
alpha = 0.9
 
# Control colour
colors = np.empty(axes + [4], dtype=np.float32)
 
colors[:] = [1, 0, 0, alpha]  # red
 
# Plot figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
 
# Voxels is used to customizations of the
# sizes, positions and colors.
ax.voxels(data, facecolors=colors)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#      10x10x10
cube = np.random.randint(0, 2, size=(3, 3, 3))

#      3D-
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#    
x, y, z = np.where(cube == 1)

#    
ax.scatter(x, y, z, c='b', marker='s')

#    
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)
ax.set_zlim(0, 3)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend(['Cube'])

#  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#      10x10x10
cube = np.random.randint(0, 2, size=(10, 10, 10))

#      
x, y, z = np.meshgrid(range(cube.shape[0]), range(cube.shape[1]), range(cube.shape[2]))

#   
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, facecolors=np.where(cube, 'b', 'w'), alpha=0.5)

#    
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend(['Cube'])

#  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#      10x10x10
cube = np.random.randint(2, size=(3, 3, 3))

#       
colors = np.empty(cube.shape, dtype=object)
colors[cube == 0] = 'white'
colors[cube == 1] = 'red'

#   ,         
fig, axs = plt.subplots(2, 2, figsize=(3, 3), subplot_kw={'projection': '3d'})

#     
axs[0, 0].voxels(cube, facecolors=colors, edgecolor='red',alpha =0.5)
axs[0, 0].set_title('Rakurs 1')

#     
axs[0, 1].voxels(np.rot90(cube, 1, axes=(0, 2)), facecolors=np.rot90(colors, 1, axes=(0, 2)), edgecolor='red',alpha =0.5)
axs[0, 1].set_title('Rakurs 2')

#     
axs[1, 0].voxels(np.rot90(cube, 1, axes=(1, 2)), facecolors=np.rot90(colors, 1, axes=(1, 2)), edgecolor='red',alpha =0.5)
axs[1, 0].set_title('Rakurs 3')

#     
axs[1, 1].voxels(np.rot90(cube, 1, axes=(0, 1)), facecolors=np.rot90(colors, 1, axes=(0, 1)), edgecolor='red',alpha =0.5)
axs[1, 1].set_title('Rakurs 4')

#  
plt.show()
```
